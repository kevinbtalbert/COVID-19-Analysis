```python
import pandas as pd
```

```python
# Import the datasets 
us_daily_df = pd.read_csv("https://github.com/COVID19Tracking/covid-tracking-data/raw/master/data/us_daily.csv")
us_current_df = pd.read_csv("https://github.com/COVID19Tracking/covid-tracking-data/raw/master/data/us_current.csv")
cdc_deaths_by_state_df = pd.read_csv("https://raw.githubusercontent.com/COVID19Tracking/covid-tracking-data/master/data/cdc_cases_deaths.csv")
```

```python
## UNDERSTAND DATA STRUCTURE
```

```python
# Print the dataframe structure for US Daily COVID-19
print(us_daily_df.columns)
```

```python
# Print the dataframe structure for US Current COVID-19
print(us_current_df.columns)
```

```python
# Print the dataframe structure for CDC's Death by States from COVID-19 Dataset
print(cdc_deaths_by_state_df.columns)
```

```python
## PROCESS DATAFRAMES AND HANDLE DATA PREPARATION
```

```python
us_daily_df['date'] = pd.to_datetime(us_daily_df['date'], infer_datetime_format=True)
us_current_df['date'] = pd.to_datetime(us_current_df['date'], infer_datetime_format=True)
cdc_deaths_by_state_df['submission_date'] = pd.to_datetime(cdc_deaths_by_state_df['submission_date'], infer_datetime_format=True)

us_daily_df['month'] = us_daily_df['date'].dt.month
us_current_df['month'] = us_current_df['date'].dt.month
cdc_deaths_by_state_df['submission_date_month'] = cdc_deaths_by_state_df['submission_date'].dt.month

us_daily_df = us_daily_df.fillna(0)
us_current_df = us_current_df.fillna(0)
cdc_deaths_by_state_df = cdc_deaths_by_state_df.fillna(0)
```
